# Example 1

Print the prime numbers from 1 to 100.

```python
# Example 1
for i in range(101):
  prime=True
  j=2
  while prime and j<i:
    if (i % (j))==0:
      prime=False
    j+=1
  if prime:
    print(i)
```

# Task 1

Modify previous example and define function for testing if the number is a prime number.

```python
# Write your code here

#Task-1
def is_prime(n):
  prime=True
  j=2
  while prime and j<n:
    if (n % (j))==0:
      prime=False
    j+=1
  return prime

def prime(n):
  for i in range(n+1):
    if is_prime(i):
      print(i)

prime(100)
```

# Task 2

Modify the previous task and define class finding the prime numbers in selected range.


```python
class prime_numbers:
  def __init__(self,n):
    self.n = n
    self.result = []

  def is_prime(self):
    prime=True
    j=2
    while prime and j<self.n:
      if (self.n % (j))==0:
        prime=False
      j+=1
    return prime

  def get_result(self):
    for i in range(self.n+1):
      if is_prime(i):
        self.result.append(i)
      
    return self.result
```

```python
obj = prime_numbers(100)
print(obj.get_result())
```

## Task 7

Normalize features in the ``iris_dataset`` into the range of $<0,1>$ using min-max norm according to the following formula:

$$ y = \frac{y_{max} - y_{min}}{x_{max}-x_{min}}(x - x_{min}) + y_{min} $$

where $x$ is raw value and $y$ is the normalised value. The $y_{min}$ and $y_{max}$ determine the output range. The $x_{max}$ and $x_{min}$ denote the input range. 

For this task $y_{max}=1$, $y_{min}=0$. The $x_{max}$ and $x_{min}$ must be obtained from the dataset, for each column separately. 

The result prints with limited precision to two decimal digits (0.00).

```python
# Write your code here
y_min = [0, 0, 0, 0]
y_max = [1, 1, 1, 1,]
x_min = np.amin(iris_features, axis=0)
x_max = np.amax(iris_features, axis=0)

print('\n',"x_min:",x_min,'\n',"x_max:",x_max,'\n',"y_min:",y_min,'\n',"y_max:",y_max)
```

```python
# Write your code here
normalized_Data = []

for row in iris_features:
  normalized_row = (((np.array(y_max) - np.array(y_min))/(np.array(x_max) - np.array(x_min))))*(np.array(row) - np.array(x_min))+np.array(y_min)
  normalized_Data.append(np.around(normalized_row, decimals=2))

normalized_Data = np.array(normalized_Data)
normalized_Data
```

```python
# Write your code here

x = normalized_Data[:,0]
y = normalized_Data[:,1]
colors = {'Iris-versicolor':'r','Iris-virginica':'g','Iris-setosa':'b'}
plt.scatter(x, y, c = iris_Dataset_enc['Label'].map(colors))
plt.show()
```

# Task 8

The sigmoid function is frequently used activation of a neuron. It is defined by

$$y=\sigma(x)=\frac{1}{1+e^{-x}}$$ 

Draw a graf of this function for $x\in <-8,8>$. 

```python
# Write your code here

import math
x = np.linspace(-8,8,100)
y = []
for i in x:
  y.append(1/(1+math.exp(-i)))

plt.plot(x,y,c = 'blue', label="$\sigma(x)$ = $\\frac{1}{1+e^{-x}}$")
plt.xlabel('x')
plt.ylabel('y')
[plt.axhline(y=i, color='grey', linestyle='dotted') for i in [0,0.5,1]]
plt.axvline(x=0, color='grey', linestyle='solid')


plt.legend(fontsize=14)
plt.show()
```

# Task 11

The Gaussian function is frequently used as an activation function of RBF neurons. It is defined by the equation

$$y=g(x)=e^{\frac{-x^2}{2\sigma^2}}$$ 

Draw a graf of this function for $x\in <-10,10>$ and $\sigma=\{0.2,0.5,1,2,5\}$. 

```python
# Write your code here
def gaussian(x,sigma):
  return np.exp((-x**2)/(2*(sigma**2)))

x = np.linspace(-10,10,100)
sigma = [0.2,0.5,1,2,5]
for i in sigma:
  y = gaussian(x, i)
  plt.plot(x,y, label='g(x,{})'.format(i))


plt.axhline(y=0, c='grey', linestyle='dotted')
plt.axvline(x=0, c='grey', linestyle='solid')
plt.xlabel('x')
plt.ylabel('y')
  
plt.legend(fontsize=14)
plt.show()
```

# Task 12

Draw response of the two input perceptron neuron defined by

$$y=\sigma(\sum_{i=0}^{1}w_ix_i + b)$$

where $w_0=10$,$w_1=10$, $b=-10$.

```python
# Write your code here

x = np.arange(0, 1, 0.01)
y = np.arange(0, 1, 0.01)
x0, x1 = np.meshgrid(x, y)

z = ((10*x0)+(10*x1)) + (-10)
sigma = 1/(1+np.exp(-z))

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.plot_surface(x0, x1, sigma,cmap='plasma')
ax.set_xlabel('X')  
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
